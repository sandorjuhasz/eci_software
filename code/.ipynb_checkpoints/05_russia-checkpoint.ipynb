```python
import pandas as pd
import glob
```

```python
lodf=[]
for f in glob.glob('../outputs/complexity_table20*.csv'):
    if f!='../outputs/complexity_table2020-2021.csv':
        lodf.append(pd.read_csv(f,sep=';'))
```

```python
df=pd.concat(lodf)
```

```python
df=df.drop_duplicates(['iso2_code','year']).sort_values('year')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df
```

```python
emigration = ['GE','ME','CY','AM','KZ','AE','RS','KZ','LT','EE','UZ']

def rus_emi_other(xx):
    if xx in emigration:
        return 'Top emigration destination'
    elif xx =='RU':
        return 'Russia'
    else:
        return 'Other'

df['is_emi'] = df['iso2_code'].map(rus_emi_other)
```

```python
df.groupby(['is_emi','year'])['eci'].mean().unstack(0).diff(1).plot()
plt.xticks([2021,2022,2023])
plt.ylabel('YoY Chg in SECI')
plt.grid()
```

```python
sns.lmplot(x='year',y='eci',hue='iso2_code',lowess=True,
           data=df[df['iso2_code'].isin(['RU','GE','ME','CY','AM','KZ','AE','RS','KZ','LT','EE','UZ'])],ci=0)
plt.xticks([2020,2021,2022,2023])
plt.yticks([0,1,2,3])
plt.grid()
plt.show()
```
