```python
import geopandas as gpd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np

# Coordinates for the largest cities in Russia
moscow_coords = (37.6173, 55.7558)
spb_coords = (30.3351, 59.9343)

# Coordinates for the destination countries
destinations = {
    'Georgia': (44.7930, 41.7151),
    'Montenegro': (19.2595, 42.7087),
    'Cyprus': (33.4299, 35.1264),
    'Armenia': (44.5126, 40.1792),
    'UAE': (54.3773, 24.4539),
    'Kazakhstan': (71.4278, 51.1694),
    'Serbia': (20.4489, 44.7866),
    'Kyrgyzstan': (74.5698, 42.8820),
    'Lithuania': (25.2797, 54.6892),
    'Estonia': (24.7535, 59.4370)
}

# Create a new plot
fig, ax = plt.subplots(figsize=(15, 10))

# Setup Basemap
m = Basemap(projection='merc', llcrnrlat=20, urcrnrlat=70, llcrnrlon=20, urcrnrlon=90, resolution='i')

# Draw map details
m.drawcountries()
m.drawcoastlines()
m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='lightgreen', lake_color='aqua')
m.drawrivers()

# Plot Moscow and St Petersburg
moscow_x, moscow_y = m(moscow_coords[0], moscow_coords[1])
spb_x, spb_y = m(spb_coords[0], spb_coords[1])
m.plot(moscow_x, moscow_y, 'ro', markersize=12)
m.plot(spb_x, spb_y, 'bo', markersize=12)

# Annotate cities
plt.text(moscow_x, moscow_y, ' Moscow', fontsize=12, ha='left', color='white', weight='bold')
plt.text(spb_x, spb_y, ' St Petersburg', fontsize=12, ha='left', color='white', weight='bold')

# Plot flows to destinations
for country, coords in destinations.items():
    dest_x, dest_y = m(coords[0], coords[1])
    m.plot(dest_x, dest_y, 'go', markersize=10)
    plt.text(dest_x, dest_y, f' {country}', fontsize=10, ha='left', color='black', weight='bold')

    # Draw curved arrows from Moscow and St Petersburg to destinations
    for start_coords in [moscow_coords, spb_coords]:
        start_x, start_y = m(start_coords[0], start_coords[1])
        m.drawgreatcircle(start_coords[0], start_coords[1], coords[0], coords[1], linewidth=2, color='red', alpha=0.6)

plt.title('Flows from Moscow and St Petersburg to Various Countries', fontsize=16)
plt.show()
```
